# Assignment1 (if)

```python
inp = input('enter grade: ')
grd = float(inp)
if grd >= 0.9:
    print("A")
elif grd >= 0.8:
    print("B")
elif grd >= 0.7:
    print("C")
elif grd >= 0.6:
    print("D")
elif grd < 0.6:
    print("F")
else:
    print("enter a valid input ")
```

# Assignment2 (for loop)

```python
#for loop to find prime numbers bw 1 and 1000
```

```python
a = 1
b = 1000
print("Prime numbers between 1 and 1000 are:")
for num in range(a, b + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```

# Assignment3 (for loop)

```python
#printing the tables of integers using nested for loops
```

```python
for num in range(1,20):
    print("multiplication table of number ",num)
    for i in range(1, 11):
        print(num, 'x', i, '=', num*i)
```
